```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
# Load Dataset
df = pd.read_csv("../Data/student_placement_prediction_dataset_2026.csv")
```

```python
df.head()
```

```python
# checking missing values
df.isnull().sum()
```

```python
# checking duplicates rows
df.duplicated().sum(0)
```

```python
df.columns
```

```python
# Drop unnecessary columns
df.drop("student_id", axis=1, inplace=True)
```

```python
df.head()
```

```python
# Converting the placement labels into numbers.
df["placement_status"] = df["placement_status"].map({
    "Placed" : 1,
    "Not Placed" : 0
})
```

```python
df.placement_status.value_counts()
```

```python
# Separating Features and Targets
x = df.drop(["placement_status", "salary_package_lpa"], axis=1)
y_class = df["placement_status"]
y_reg = df["salary_package_lpa"]
```

```python
x.select_dtypes(include="object").columns
```

```python
# Encoding of Categorical Features
x = pd.get_dummies(x, drop_first=True)
```

```python
x.head()
```

```python
feature_columns = x.columns
```

```python
# Saving feature names
import pickle

pickle.dump(feature_columns, open("../models/feature_columns.pkl", "wb"))
```

```python
# Scaling Numerical Features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# Saving the scaler
pickle.dump(scaler, open("../models/scaler.pkl", "wb"))
```

```python
# Train-Test Split (Classification)

from sklearn.model_selection import train_test_split
x_train_cls, x_test_cls, y_train_cls, y_test_cls = train_test_split(x_scaled, y_class, test_size=0.2, random_state=42, stratify=y_class)
```

```python
# Train-Test Split (Regression)
placed_df = df[df["placement_status"] == 1]
```

```python
# Separation of features and targets
x_reg = placed_df.drop(["placement_status", "salary_package_lpa"], axis=1)
y_reg = placed_df["salary_package_lpa"]
```

```python
# Encoding of Categorical Features
x_reg = pd.get_dummies(x_reg, drop_first=True)
```

```python
x_reg = x_reg.reindex(columns=feature_columns, fill_value=0)
```

```python
x_reg.head()
```

```python
x_reg = scaler.transform(x_reg)
```

```python
x_train_reg, x_test_reg, y_train_reg, y_test_reg = train_test_split(
    x_reg,
    y_reg,
    test_size=0.2,
    random_state=42
)
```

```python
print(x_train_cls.shape)
print(x_test_cls.shape)

print(x_train_reg.shape)
print(x_test_reg.shape)
```

```python
# Save Classification Data
pickle.dump(x_train_cls, open("../models/x_train_cls.pkl", "wb"))
pickle.dump(x_test_cls, open("../models/x_test_cls.pkl", "wb"))
pickle.dump(y_train_cls, open("../models/y_train_cls.pkl", "wb"))
pickle.dump(y_test_cls, open("../models/y_test_cls.pkl", "wb"))

# Save Regression Data
pickle.dump(x_train_reg, open("../models/x_train_reg.pkl", "wb"))
pickle.dump(x_test_reg, open("../models/x_test_reg.pkl", "wb"))
pickle.dump(y_train_reg, open("../models/y_train_reg.pkl", "wb"))
pickle.dump(y_test_reg, open("../models/y_test_reg.pkl", "wb"))
```
